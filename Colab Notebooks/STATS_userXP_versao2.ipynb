```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from matplotlib.backends.backend_pdf import PdfPages

def compute_stats_and_plot(df, title, output_path):
    """
    Computes statistics and generates plots for the 'Distance' column,
    grouped by 'Method', and saves them to a PDF file.

    Args:
        df: pandas DataFrame containing the data.
        title: Title for the plots and the PDF file.
        output_path: Path to save the PDF file.
    """

    # Get unique methods and assign colors
    methods = df['METHOD'].unique()
    # Define specific colors for metricSync and pulseSync
    method_colors = {
        'metricSync': 'gold',
        'pulseSync': 'magenta',
        'beatSync': 'lime'
    }
    # Create a color map for other methods if any
    other_methods = [m for m in methods if m not in method_colors]
    cmap = plt.cm.get_cmap('viridis', len(other_methods))
    color_map = {method: color for method, color in zip(other_methods, cmap.colors)}
    # Combine method_colors and color_map
    color_map.update(method_colors)

    with PdfPages(output_path + '/' + title + '_stats.pdf') as pdf:
        for method in methods:
            data = df[df['METHOD'] == method]['DISTANCE']
            color = color_map[method]

            # Compute statistics
            mean = np.mean(data)
            median = np.median(data)
            mode = stats.mode(data).mode
            mode = mode[0] if isinstance(mode, np.ndarray) and mode.size > 0 else mode
            std = np.std(data)
            variance = np.var(data)
            minimum = np.min(data)
            maximum = np.max(data)
            range_val = maximum - minimum
            q1, q2, q3 = np.percentile(data, [25, 50, 75])

            # Print statistics (optional)
            print(f"Statistics for {method}:")
            print(f"  Mean: {mean}")
            print(f"  Median: {median}")
            print(f"  Mode: {mode}")
            print(f"  Standard Deviation: {std}")
            print(f"  Variance: {variance}")
            print(f"  Minimum: {minimum}")
            print(f"  Maximum: {maximum}")
            print(f"  Range: {range_val}")
            print(f"  Quartiles: Q1={q1}, Q2={q2}, Q3={q3}")

            # Create plots
            fig, axes = plt.subplots(1, 2, figsize=(12, 6))

            # Histogram
            axes[0].hist(data, color=color, bins='auto')
            axes[0].set_title(f"Histogram of Distance for {method}")
            axes[0].set_xlabel("Distance")
            axes[0].set_ylabel("Frequency")

            # Data Distribution Plot (e.g., Boxplot)
            axes[1].boxplot(data, vert=False, patch_artist=True,
                            boxprops=dict(facecolor=color))
            axes[1].set_title(f"Data Distribution of Distance for {method}")
            axes[1].set_xlabel("Distance")

            plt.tight_layout()
            pdf.savefig(fig)  # Save figure to PDF
            plt.close(fig)  # Close figure to free memory

# Call the function for both DataFrames with the output path
output_path = "/content/drive/MyDrive/Colab Notebooks/U_EXP/versao2/"
compute_stats_and_plot(dfM, 'dfM', output_path)
compute_stats_and_plot(dfP, 'dfP', output_path)
compute_stats_and_plot(dfB, 'dfB', output_path)

print("Statistics and plots saved to PDF files in the specified folder.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from matplotlib.backends.backend_pdf import PdfPages

def compute_stats_and_plot(df, title, output_path):
    """
    Computes statistics and generates plots for the 'Distance' column,
    grouped by 'Method', and saves them to a PDF file.

    Args:
        df: pandas DataFrame containing the data.
        title: Title for the plots and the PDF file.
        output_path: Path to save the PDF file.
    """

    # Get unique methods and assign colors
    methods = df['METHOD'].unique()
    # Define specific colors for metricSync and pulseSync
    method_colors = {
        'metricSync': 'gold',
        'pulseSync': 'magenta',
        'beatSync': 'lime'
    }
    # Create a color map for other methods if any
    other_methods = [m for m in methods if m not in method_colors]
    cmap = plt.cm.get_cmap('viridis', len(other_methods))
    color_map = {method: color for method, color in zip(other_methods, cmap.colors)}
    # Combine method_colors and color_map
    color_map.update(method_colors)

    with PdfPages(output_path + '/' + exp_title + title + '_stats.pdf') as pdf:
        for method in methods:
            data = df[df['METHOD'] == method]['DISTANCE']
            color = color_map[method]

            # Compute statistics
            mean = np.mean(data)
            median = np.median(data)
            mode = stats.mode(data).mode
            mode = mode[0] if isinstance(mode, np.ndarray) and mode.size > 0 else mode
            std = np.std(data)
            variance = np.var(data)
            minimum = np.min(data)
            maximum = np.max(data)
            range_val = maximum - minimum
            q1, q2, q3 = np.percentile(data, [25, 50, 75])

            # Print statistics (optional)
            print(f"Statistics for {method}:")
            print(f"  Mean: {mean}")
            print(f"  Median: {median}")
            print(f"  Mode: {mode}")
            print(f"  Standard Deviation: {std}")
            print(f"  Variance: {variance}")
            print(f"  Minimum: {minimum}")
            print(f"  Maximum: {maximum}")
            print(f"  Range: {range_val}")
            print(f"  Quartiles: Q1={q1}, Q2={q2}, Q3={q3}")

            # Create plots
            fig, axes = plt.subplots(1, 2, figsize=(12, 6))

            # Histogram
            axes[0].hist(data, color=color, bins='auto')
            axes[0].set_title(f"Histogram of Distance for {method}")
            axes[0].set_xlabel("Distance")
            axes[0].set_ylabel("Frequency")
            # Add vertical lines for quartiles on histogram
            axes[0].axvline(q1, color='red', linestyle='--', label='Q1')
            axes[0].axvline(q2, color='blue', linestyle='-', label='Q2 (Median)')
            axes[0].axvline(q3, color='turquoise', linestyle='--', label='Q3')
            axes[0].legend()

            # Data Distribution Plot (e.g., Boxplot)
            axes[1].boxplot(data, vert=False, patch_artist=True,
                            boxprops=dict(facecolor=color), showfliers=False)  # showfliers=False to hide outliers
            axes[1].set_title(f"Data Distribution of Distance for {method}")
            axes[1].set_xlabel("Distance")
            # Add text labels for quartiles on boxplot
            axes[1].text(q1, 1.1, f"Q1: {q1:.2f}", color='red', ha='center', va='center')
            axes[1].text(q2, 1.2, f"Q2: {q2:.2f}", color='blue', ha='center', va='center')
            axes[1].text(q3, 1.3, f"Q3: {q3:.2f}", color='turquoise', ha='center', va='center')


            plt.tight_layout()
            pdf.savefig(fig)  # Save figure to PDF
            plt.close(fig)  # Close figure to free memory

# Call the function for both DataFrames with the output path
output_path = "/content/drive/MyDrive/Colab Notebooks/U_EXP/versao2"
compute_stats_and_plot(dfM, 'dfM', output_path)
compute_stats_and_plot(dfP, 'dfP', output_path)
compute_stats_and_plot(dfB, 'dfB', output_path)

print("Statistics and plots saved to PDF files in the specified folder.")
```
