```python
import pandas as pd

file_path = "/content/drive/MyDrive/Colab Notebooks/U_EXP/Affinity Gen Results.csv"

df = pd.read_csv(file_path)

display(df)

exp_title = "U_EXP_00"
sort_by = "3/4"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'df' and the relevant columns are 'Distance' and 'Method'

# Group data by 'Method'
grouped_data = df.groupby('Method')['Distance']

# Calculate statistics for each group
statistics = grouped_data.agg(['mean', 'median', 'std', 'var',
                                 lambda x: x.mode()[0],  # Mode calculation using the first element
                                 'min', 'max', lambda x: x.quantile(0.25),  # Q1 calculation
                                 lambda x: x.quantile(0.75)])  # Q3 calculation

# Rename columns for clarity
statistics.columns = ['Mean', 'Median', 'Std Dev', 'Variance', 'Mode',
                      'Min', 'Max', 'Q1', 'Q3']

# Display statistics
display(statistics)

# Plotting with different colors for each group
colors = sns.color_palette('husl', n_colors=len(grouped_data))  # Generate a color palette

fig, axes = plt.subplots(len(grouped_data), 2, figsize=(15, 5 * len(grouped_data)))
fig.suptitle('Distribution and Histogram of Distance by Method', fontsize=16)

for i, (method, data) in enumerate(grouped_data):
    # Distribution plot
    sns.kdeplot(data, color=colors[i], ax=axes[i, 0], fill=True)
    axes[i, 0].set_title(f'Distribution of Distance for {method}')
    axes[i, 0].set_xlabel('Distance')
    axes[i, 0].set_ylabel('Density')

    # Histogram
    axes[i, 1].hist(data, color=colors[i], bins=20)
    axes[i, 1].set_title(f'Histogram of Distance for {method}')
    axes[i, 1].set_xlabel('Distance')
    axes[i, 1].set_ylabel('Frequency')

plt.tight_layout(rect=[0, 0, 1, 0.98])  # Adjust layout to prevent overlapping
plt.show()
```
