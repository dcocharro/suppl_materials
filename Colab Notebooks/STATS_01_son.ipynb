```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from matplotlib.backends.backend_pdf import PdfPages

def compute_stats_and_plot(df, title):
  """Computes statistics and generates plots for the 'Distance' column.

  Args:
    df: pandas DataFrame containing the data.
    title: Title for the plots.
  """

  # Get unique methods and assign colors
  methods = df['Method'].unique()
  colors = plt.cm.get_cmap('viridis', len(methods))

  with PdfPages(title + '_stats.pdf') as pdf:
    for method, color in zip(methods, colors):
      data = df[df['Method'] == method]['Distance']

      # Compute statistics
      mean = np.mean(data)
      median = np.median(data)
      mode = stats.mode(data).mode[0]
      std = np.std(data)
      variance = np.var(data)
      minimum = np.min(data)
      maximum = np.max(data)
      range_val = maximum - minimum
      q1, q2, q3 = np.percentile(data, [25, 50, 75])

      # Print statistics
      print(f"Statistics for {method}:")
      print(f"  Mean: {mean}")
      print(f"  Median: {median}")
      print(f"  Mode: {mode}")
      print(f"  Standard Deviation: {std}")
      print(f"  Variance: {variance}")
      print(f"  Minimum: {minimum}")
      print(f"  Maximum: {maximum}")
      print(f"  Range: {range_val}")
      print(f"  Quartiles: Q1={q1}, Q2={q2}, Q3={q3}")

      # Create plots
      fig, axes = plt.subplots(1, 2, figsize=(12, 6))

      # Histogram
      axes[0].hist(data, color=color, bins='auto')
      axes[0].set_title(f"Histogram of Distance for {method}")
      axes[0].set_xlabel("Distance")
      axes[0].set_ylabel("Frequency")

      # Data Distribution Plot (e.g., Boxplot)
      axes[1].boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor=color))
      axes[1].set_title(f"Data Distribution of Distance for {method}")
      axes[1].set_xlabel("Distance")

      plt.tight_layout()
      pdf.savefig(fig)  # Save figure to PDF
      plt.close(fig)  # Close figure to free memory


# Call the function for both DataFrames
compute_stats_and_plot(dfM, 'dfM')
compute_stats_and_plot(dfP, 'dfP')

print("Statistics and plots saved to dfM_stats.pdf and dfP_stats.pdf")
```
