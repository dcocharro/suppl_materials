```python
import pandas as pd
import numpy as np

def compute_dft(signal):
    try:
        # 1. Convert to numerical list:
        signal = [int(i) for i in signal.split()]

        # 2. Perform DFT calculation:
        dft = np.fft.fft(signal)

        # 3. Normalize by DC energy:
        dc_energy = np.real(dft[0])  # Get real part of DC component
        if dc_energy != 0:
            dft = dft / dc_energy  # Normalize if DC energy is not zero
        else:
            print("Warning: DC energy is zero. Normalization skipped.")

        # 4. Exclude DC component and symmetric part:
        dft = dft[1:len(dft)//2]  # Exclude DC and symmetric part

        # 5. Calculate magnitudes and phases:
        magnitudes = np.abs(dft)
        phases = np.angle(dft)

        return dft.tolist(), magnitudes.tolist(), phases.tolist()  # Return as lists

    except (ValueError, TypeError) as e:
        print(f"Error computing DFT for {signal}: {e}")
        return [np.nan], [np.nan], [np.nan]  # Return list of NaNs on error

# ... (your code to load the DataFrame df) ...

# Assuming 'Rhythm Binary' column contains strings of binary values
df[['DFT', 'DFT Magnitude', 'DFT Phase']] = df['Rhythm Binary'].apply(compute_dft).apply(pd.Series)
display(df)
```

```python
from sklearn.manifold import MDS
import matplotlib.pyplot as plt

# 1. Compute pairwise distances using Euclidean distance
distances = pd.DataFrame(np.zeros((len(df), len(df)))) # Initialize distance matrix
for i in range(len(df)):
  for j in range(i + 1, len(df)):
    # Convert lists to NumPy arrays before subtraction
    dist = np.linalg.norm(np.array(df.iloc[i]['DFT Magnitude']) - np.array(df.iloc[j]['DFT Magnitude']))
    distances.iloc[i, j] = dist
    distances.iloc[j, i] = dist  # Ensure symmetry

# 2. Apply non-metric MDS
embedding = MDS(n_components=2, dissimilarity='precomputed', metric=False)
transformed_data = embedding.fit_transform(distances)

# 3. Visualize the results
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], alpha=0)  # alpha=0 for invisibility
for i in range(len(df)):
    plt.text(transformed_data[i, 0], transformed_data[i, 1], str(df.index[i]),
             fontsize=10, ha='center', va='center')
plt.title('Non-metric MDS Visualization of DFT Magnitudes (Euclidean Distance)')
plt.xlabel('MDS Dimension 1')
plt.ylabel('MDS Dimension 2')
plt.show()
```

```python
import pandas as pd
import numpy as np

# ... (your existing code) ...

def compute_dft_norm(dft_values):
    """Computes the norm (magnitude) of a DFT vector."""
    try:
        dft_array = np.array(dft_values) # Convert to numpy array
        norm = np.linalg.norm(dft_array) # Calculate norm using numpy
        return norm
    except (ValueError, TypeError) as e:
        print(f"Error in compute_dft_norm for entry: {dft_values}: {e}")
        return np.nan

# Assuming 'DFT' column contains the DFT values (as lists)
df['DFT Norm'] = df['DFT'].apply(compute_dft_norm)

correlation = df['DFT Norm'].corr(df['Perceptual Complexity Shmulevich and Povel'])
print(f"Correlation between DFT Norm and Perceptual Complexity: {correlation}")
display(df)
```
